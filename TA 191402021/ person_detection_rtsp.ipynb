```python
import cv2
import time

# ===== RTSP CAMERA URL =====
RTSP_URL = "rtsp://admin:qwerty170845@10.1.15.58:554/mode=real&idc=1&ids=1"

# ===== MODEL FILES =====
PROTOTXT = "deploy.prototxt"
MODEL = "mobilenet_iter_73000.caffemodel"

# ===== LOAD MODEL =====
print("[INFO] Loading SSD MobileNet model...")
net = cv2.dnn.readNetFromCaffe(PROTOTXT, MODEL)

# Optional: Enable CUDA (if OpenCV built with CUDA)
# net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
# net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)

# ===== CONNECT TO RTSP STREAM =====
print(f"[INFO] Connecting to RTSP stream: {RTSP_URL}")
cap = cv2.VideoCapture(RTSP_URL)

if not cap.isOpened():
    print("[ERROR] Cannot open RTSP stream. Check camera connection.")
    exit()

# ===== CLASS NAMES =====
CLASSES = ["background", "aeroplane", "bicycle", "bird", "boat",
           "bottle", "bus", "car", "cat", "chair", "cow", "diningtable",
           "dog", "horse", "motorbike", "person", "pottedplant", "sheep",
           "sofa", "train", "tvmonitor"]

CONF_THRESHOLD = 0.5  # Confidence threshold
fps_start_time = time.time()
frame_count = 0
fps = 0.0

# ===== MAIN LOOP =====
while True:
    ret, frame = cap.read()
    if not ret:
        print("[WARNING] Frame not received, reconnecting...")
        time.sleep(1)
        cap.release()
        cap = cv2.VideoCapture(RTSP_URL)
        continue

    frame_count += 1
    (h, w) = frame.shape[:2]

    # Prepare input blob
    blob = cv2.dnn.blobFromImage(cv2.resize(frame, (300, 300)),
                                 0.007843, (300, 300), 127.5)
    net.setInput(blob)
    detections = net.forward()

    # Loop over detections
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]

        if confidence > CONF_THRESHOLD:
            idx = int(detections[0, 0, i, 1])
            label = CLASSES[idx]

            # Only detect persons
            if label == "person":
                box = detections[0, 0, i, 3:7] * [w, h, w, h]
                (startX, startY, endX, endY) = box.astype("int")

                # Draw bounding box
                cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)
                text = f"Person: {confidence * 100:.1f}%"
                y = startY - 15 if startY - 15 > 15 else startY + 15
                cv2.putText(frame, text, (startX, y),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Calculate FPS
    if frame_count >= 10:
        fps_end_time = time.time()
        fps = frame_count / (fps_end_time - fps_start_time)
        fps_start_time = fps_end_time
        frame_count = 0

    cv2.putText(frame, f"FPS: {fps:.2f}", (10, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

    # Show frame
    cv2.imshow("Real-Time Person Detection (RTSP)", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("[INFO] Exiting...")
        break

cap.release()
cv2.destroyAllWindows()
```
